```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image


def display(display_list):
  plt.figure(figsize = (15, 15))

  title = ['Input Image', 'Mask']

  for i in range(len(display_list)):
    plt.subplot(1, len(display_list), i+1)
    plt.title(title[i])
    plt.imshow(np.asarray(Image.open(display_list[i])))
    plt.axis('off')
    print("Dimensions images:", np.asarray(Image.open(display_list[i])).shape, "et Masques", np.asarray(Image.open(display_list[i])).shape)

  plt.show()



display_list = ['/content/drive/MyDrive/dermato/data/processed/train/images/image_0.png',
                '/content/drive/MyDrive/dermato/data/processed/train/masks/mask_0.png']

display(display_list)
```
